```python
from langchain_openai import OpenAIEmbeddings
from langchain.vectorstores import Chroma
```

```python
from langchain.schema import Document
doc1 = Document(
    page_content="Virat Kohli",
    metadata={"team": "RCB"}
)

doc2 = Document(
    page_content="Rohit Sharma",
    metadata={"team": "MI"}
)

doc3 = Document(
    page_content="MS Dhoni",
    metadata={"team": "CSK"}
)
```

```python
docs = [doc1, doc2, doc3]
```

```python
vector_store = Chroma(
    embedding_function=OpenAIEmbeddings(),
    persist_directory='chroma_db',
    collection_name='sample'
)
```

```python
vector_store.add_documents(docs)
```

```python
vector_store.add_documents(docs)
```

```python
vector_store.get(include=["embeddings", "documents", "metadatas"])
```

```python
vector_store.similarity_search(
    query="Virat Kohli",
    k=1
)
```

```python
vector_store.similarity_search_with_score(
    query="MS",
    k=2
)
```

```python
vector_store.similarity_search_with_score(
    query="",
    k=2,
    filter={"team": "CSK"}
)
```

```python
update_doc1 = Document(
    page_content="Virat Kohli 2",
    metadata={"team": "RCB"}
)
vector_store.update_document(document_id="ID", document=update_doc1)
```

```python
vector_store.delete(ids=['ID'])
```
